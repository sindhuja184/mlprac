```python
from sklearn.tree import DecisionTreeClassifier 
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import LabelEncoder

label = LabelEncoder()

df = pd.read_csv(r"C:\Users\sindh\OneDrive\Desktop\datasets\iris.csv")
df['Species'] = label.fit_transform(df['Species'])


dt = DecisionTreeClassifier()

params_grid = {
    'criterion' : ['gini', 'entropy'],
    'max_depth': [3, 5, 10, 20],
    'splitter' : ['best', 'random']
}

grid_search = GridSearchCV(estimator = dt, param_grid = params_grid, cv =5, scoring = 'accuracy')

X = df.drop(['Species'], axis =1)
y = df['Species']

grid_search.fit(X, y)
print("Best Parameters", grid_search.best_params_)

best = grid_search.best_estimator_
y_pred = best.predict(X)

acc = accuracy_score(y, y_pred)
print("Accuracy is ", acc)
```

```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import mode
from sklearn.preprocessing import LabelEncoder
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

df = pd.read_csv(r"C:\Users\sindh\OneDrive\Desktop\datasets\iris.csv")
df = df.drop(['Id'], axis = 1)

label = LabelEncoder()
df['Species'] = label.fit_transform(df['Species'])

kmeans = KMeans(n_clusters = 3)
X = df.drop(['Species'], axis = 1)
y = df['Species']


y_pred = kmeans.fit_predict(X)
print(y_pred)

df['cal'] = y_pred

df0 = df[df.cal == 0]
df1 = df[df.cal == 1]
df2 = df[df.cal == 2]

##Plotting the graph
plt.scatter(df0['SepalLengthCm'], df0['SepalWidthCm'], color = 'orange')
plt.scatter(df1['SepalLengthCm'], df1['SepalWidthCm'], color = 'pink')
plt.scatter(df2['SepalLengthCm'], df2['SepalWidthCm'], color = 'blue')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:,1], color = 'black', marker = '*')
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')

labels = np.zeros_like(y_pred)
for i in range(3):
    mask = (y_pred == i)
    labels[mask] = mode(y[mask])[0]

true = 0
for i, j in zip(labels, y):
    if i == j:
        true += 1
print("Accuracy is ", true / len(y))
```

# SINGLE LINEAR REGRESSION

```python
class SingleLinearRegression:
    def __init__(self):
        self.m = 0
        self.c = 0
    def fit(self, X, y):
        X = X.values
        y = y.values
        xm, ym = np.mean(X), np.mean(y)
        self.m = np.sum((X - xm) * (y - ym)) / np.sum((X - xm) ** 2)
        self.c = ym - self.m * xm
    def predict(self, X):
        return X * self.m + self.c

def plot(X, y, y_pred):
    X, y, y_pred = X.values, y.values, y_pred.values
    plt.scatter(X, y, color = 'pink')
    plt.plot(X, y_pred, color = 'blue')
    plt.xlabel('year')
    plt.ylabel('income')

df = pd.read_csv(r"C:\Users\sindh\OneDrive\Desktop\datasets\report.csv")
model = SingleLinearRegression()
X = df['year']
y = df['income']

model.fit(X, y)
y_pred = model.predict(X)

from sklearn.metrics import mean_squared_error
print("The mean squaared error is ", mean_squared_error(y, y_pred))
from sklearn.metrics import r2_score
print("The r squared is ", r2_score(y, y_pred))
plot(X, y, y_pred)
```

# SINGLE LAYER PERCEPTRON

```python
class SingleLayerPerceptron:
    def __init__(self, lr = 0.01, epoch = 1000):
        self.lr = lr
        self.epoch = epoch
    def fit(self, X, y):
        X = X.values
        X_shape, X_features = X.shape
        self.w = np.zeros(X_features)
        self.bias = 0

        for _ in range(self.epoch):
            for t, target in zip(X, y):
                update = self.lr * (target - self.predict(t))
                self.w = self.w + update * t
                self.bias =self.bias + update
    def predict(self, X):
        return np.where(np.dot(X,self.w) + self.bias > 0, 1, 0)
```

```python
import pandas as pd 
import numpy as np

df = pd.read_csv(r"C:\Users\sindh\OneDrive\Desktop\datasets\iris.csv")
df.Species = df.Species.replace({'Iris-setosa' :0, 'Iris-versicolor': 1, 'Iris-virginica' : 2})
df = df[(df.Species == 1) | (df.Species == 0)]
X = df[['SepalLengthCm', 'SepalWidthCm']]
y = df['Species']
l = LabelEncoder()
y = l.fit_transform(y)
model = SingleLayerPerceptron()
model.fit(X, y)
y_pred = model.predict(X)
```

# CANDIDATE ELIMINATION

```python
import numpy as np
import pandas as pd

df = pd.read_csv(r"C:\Users\sindh\OneDrive\Desktop\datasets\playground.csv")
df = df.drop(['sno'], axis = 1)
df
```

```python
data = df.values
n = len(data[0])
n -= 1
print(n)
S = ['pi'] * n
G = [['?'] * n for _ in range(n)]

isFirst = True

for row in data:
    if row[-1] == 'yes':
        if isFirst == True:
            for i in range(n):
                S[i] = row[i]
            isFirst = False
        else:
            for i in range(n):
                if S[i] != row[i]:
                    S[i] = '?'
    else:
        for i in range(n):
            if S[i] != row[i] and S[i] != '?':
                G[i][i] = S[i]

G_ans = []
for i in range(n):
    if G[i][i] != '?' and S[i] == G[i][i]:
        G_ans.append(G[i])
print("Specific hypothsis ", S)
print("General Hypothesis ", G_ans)
```
